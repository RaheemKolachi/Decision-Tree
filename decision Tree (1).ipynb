```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

from sklearn import tree
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("diabetes1.csv")
x=data.iloc[:,:-1]
y=data.iloc[:,-1:]

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
```

```python
dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
pred=dt.predict(xtest)
confusion_matrix(ytest,pred)
accuracy_score(ytest,pred)
```

```python
plt.figure(figsize=(25,20))
a=tree.plot_tree(dt,
                  filled=True)
```

```python
rep=classification_report(ytest, pred)
print(rep)
```

```python
gin= DecisionTreeClassifier(criterion = "gini",
            random_state = 100,max_depth=3)
gin.fit(xtrain, ytrain)

prg=gin.predict(xtest)

confusion_matrix(ytest,prg)
accuracy_score(ytest,prg)
```

```python
fig2 = plt.figure(figsize=(25,20))
_=tree.plot_tree(gin,feature_names=x.columns,
                  filled=True)
```
